```python
#checking for data inconsistencies
def check_missing_duplicates(data):
    na_present, duplicates_present = False, False
    if data.isna().sum().values.any():
        na_present = True
    if data.duplicated().sum() != 0:
        duplicates_present = True
    return na_present, duplicates_present
```

```python
#resolving data inconsistencies
na_present, duplicates_present = check_missing_duplicates(data)
if na_present:
    data.dropna(inplace=True)
if duplicates_present:
    data.drop_duplicates(inplace=True)
```

```python
na_present, duplicates_present = check_missing_duplicates(data)
na_present, duplicates_present
```

For a text classification problem, we could perform the following cleaning tasks:
<li> Lowercasing: reduces the number of distinct tokens </li>
<li> Removing punctuation: punctuations usually doesn't the meaning of the text in terms of classification </li>
<li> Removing numbers: from a perspective of text classification, the numbers do not affect much </li>
<li> Removing whitespaces </li>
<li> Removing stop words: avoid adding noise to the model due to the occurrance of frequent words </li>
<li> Removing html tags </li>

```python
def clean_text(text):
    #removing html tags
    try:
        text = BeautifulSoup(text, 'html.parser').get_text()
    except:
        text = re.sub(r'<.?*>', text)
    
    #lowercasing
    text = text.lower()
    
    #removing punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    #removing numbers and extra whitespaces
    text = re.sub(r'\d+', '', text.strip())
    
    #removing stop words
    stop_words = set(stopwords.words('english'))
    text = text.split()
    text = [word for word in text if word not in stop_words]
    
    return ' '.join(text)
```

```python
data['tweet'] = data['tweet'].apply(clean_text)
```

# Data Exploration from an NLP Perspective:
Understanding the characteristics of the data will help us choose whether to go along with a basic MLP model(token based model) or sequence models.

Let us collect key metrics such as:
<li> Number of samples </li>
<li> Number of classes </li>
<li> Number of samples per class to check for any data imbalance </li>
<li> Number of words per sample - this is a key metric that helps us think about numerical representation of the tokens </li>
<li> Frequence distribution of words </li>
<li> Distribution of sample length </li>

```python
def get_key_metrics(data):
    samples_length = len(data)
    print('Number of Samples present after text cleaning is', samples_length)
    num_classes = data['emotion'].nunique()
    print('Number of classes present is ', num_classes)
    num_words_per_sample_median = np.median([len(each.split()) for each in data['tweet']])
    num_words_per_sample_average = np.mean([len(each.split()) for each in data['tweet']])
    print('Number of words per sample( mean, median) is ', (num_words_per_sample_average
                                                                           , num_words_per_sample_median))
    num_samples_per_class = data['emotion'].value_counts()
    num_samples_per_class.plot(kind='bar')
    plt.title('Number of samples per class')
    plt.xlabel('Emotion')
    plt.ylabel('Number of samples')
    plt.grid(axis='y')
    plt.show()
    
    #frequency distribution of words
    kwargs = {
        'analyzer': 'word'
    }
    vectorizer = CountVectorizer(**kwargs)
    X = vectorizer.fit_transform(data['tweet'])
    all_ngrams = list(vectorizer.get_feature_names_out())
    all_counts = X.sum(axis=0).tolist()[0]
    
    #sort n-grams based on their count
    all_counts, all_ngrams = zip(*[(c,n) for c,n in sorted(zip(all_counts,all_ngrams), reverse=True)])
    n_grams_top_50 = all_ngrams[:50]
    counts_top_50 = all_counts[:50]
    idx = np.arange(len(n_grams_top_50))
    plt.figure(figsize=(12,10))
    plt.bar(idx, counts_top_50, width=0.8, color='b')
    plt.xlabel('N-grams')
    plt.ylabel('Frequencies')
    plt.title('Frequency distribution of n-grams')
    plt.xticks(idx, n_grams_top_50, rotation=45)
    plt.show()
    
    #distribution of sample length
    plt.hist([len(s.split()) for s in data['tweet']])
    plt.xlabel('Sample lengths')
    plt.ylabel('Frequency')
    plt.title('Distribution of sample lengths')
    plt.show()
    
```

```python
get_key_metrics(data)
```

```python
#let us apply lemmatization
lemmatizer = WordNetLemmatizer()
def lemmatize_texts(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

data['tweet'] = data['tweet'].apply(lemmatize_texts)
```

```python
get_key_metrics(data)
```

# Split into Training and Validation datasets

```python
def split_data(data, test_size=0.2, random_state=42):
    train_data, validation_data = train_test_split(data, test_size=test_size, random_state=random_state, stratify=data['emotion'])
    print(train_data)
    print(validation_data)
    return train_data, validation_data
```

```python
train_data, validation_data = split_data(data)
len(train_data), len(validation_data)
```

# Tokenization and building sequences

```python
num_words = 2000
oov_token = '<OOV>'
max_length = 500
padding = 'pre'
truncation = 'pre'

tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=num_words, oov_token=oov_token)

tokenizer.fit_on_texts(train_data['tweet'])

train_sequences = tokenizer.texts_to_sequences(train_data['tweet'])
valid_sequences = tokenizer.texts_to_sequences(validation_data['tweet'])

train_padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(train_sequences, maxlen=max_length, padding=padding, truncating=truncation)
valid_padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(valid_sequences, maxlen=max_length, padding=padding, truncating=truncation)

label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_data['emotion'])
valid_labels = label_encoder.fit_transform(validation_data['emotion'])

train_labels = np.array(train_labels)
valid_labels = np.array(valid_labels)
```

```python
print('First five training sequences')
print(train_padded_sequences[:5])
print(train_labels[:5])
```

```python
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)
    
    plt.figure(figsize=(14, 5))
    
    plt.subplot(1,2,1)
    plt.plot(epochs, loss, 'b', label='Training Loss')
    plt.plot(epochs, val_loss, 'r', label='Validation Loss')
    plt.title('Training vs Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(epochs, acc, 'b', label='Training Accuracy')
    plt.plot(epochs, val_acc, 'r', label='Validation Accuracy')
    plt.title('Training vs Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy %')
    plt.legend()
    
    plt.show()
```

```python
tf.keras.backend.clear_session()
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(num_words, output_dim=32, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(train_padded_sequences, train_labels, epochs=100, validation_data=(valid_padded_sequences, valid_labels))
```

```python
plot_training_history(history)
```

The above loss plot is a clear indication that the model is being overfit. Let's try adding some dropout layers and L2 regularization to a dense layer's kernel to achieve regularization. (adding dropout layers and choosing l2 constant took too much iteration to get the optimal value shown below)

```python
tf.keras.backend.clear_session()
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(num_words, output_dim=32, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu', kernel_regularizer=l2(0.01)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(train_padded_sequences, train_labels, epochs=100, validation_data=(valid_padded_sequences, valid_labels))

plot_training_history(history)
```

After adding dropout layer and L2 regularization, the model seems to generalize more but the large fluctuations in the validation loss might be an indication of a large learning rate. Let's try optimizing the learning rate below.

```python
def plot_loss_against_learning_rate(history, initial_lr, epochs):
    lrs = initial_lr * (10 ** (np.arange(epochs) / 20))
    plt.figure(figsize=(8, 5))
    plt.semilogx(lrs, history.history['loss'])
    plt.grid(True)
    plt.tick_params('both', length=10, width=1, which='both')
    plt.show()
```

```python
tf.keras.backend.clear_session()
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(num_words, output_dim=32, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu', kernel_regularizer=l2(0.01)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])

optimizer = tf.keras.optimizers.Adam()
lr_schedule = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-5 * 10**(epoch/20))

model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

#early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, min_delta=0.01, restore_best_weights=True)

history = model.fit(train_padded_sequences, train_labels, epochs=100,
                    validation_data=(valid_padded_sequences, valid_labels), callbacks=[lr_schedule])

plot_loss_against_learning_rate(history, epochs=100, initial_lr=1e-5)

#plot_training_history(history)
```

The above is a plot of learning rate against training loss. The training loss is low and stable at around 10-3. Let's try using this learning rate as our base and use a learning rate scheduler to see how our model performs.

```python
tf.keras.backend.clear_session()
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(num_words, output_dim=32, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu', kernel_regularizer=l2(0.01)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])

optimizer = tf.keras.optimizers.Adam()
initial_lr = 0.001
decay_rate = 0.80
decay_steps = 25
epochs = 500

def step_decay_schedule(epoch, lr):
    if epoch % decay_steps == 0 and epoch != 0:
        return lr * decay_rate
    return lr

lr_schedule = tf.keras.callbacks.LearningRateScheduler(step_decay_schedule)

model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

#early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, min_delta=0.01, restore_best_weights=True)

history = model.fit(train_padded_sequences, train_labels, epochs=epochs,
                    validation_data=(valid_padded_sequences, valid_labels), callbacks=[lr_schedule])

plot_loss_against_learning_rate(history, epochs=epochs, initial_lr=initial_lr)

plot_training_history(history)
```

While we were able to improve the accuracy up from 66% to 70% by dynamically decreasing the learning rate for every 20 epoch, the above graph still shows over-fitting and further needs to be validated.
